```python
product_link4=[]
```

```python
link=driver.find_elements(By.XPATH,'//a[@class="rilrtl-products-list__link desktop"]')
for i in link:
    try:
        product_link4.append(i.get_attribute('href'))
    except Exception as e:
        pass
```

```python
len(product_link4)
```

```python
ms_p_price = []
ms_b_name = []
ms_p_code = []
ms_category = []
ms_p_type = []
ms_p_name =[]

ms_col1 = []
ms_col2 = []
ms_col3 = []
ms_col4 = []

ms_m_type = []
ms_reviews = []

ms_occ = []
ms_r_time = []
ms_c_ori = []
ms_size1 = []
ms_size2 = []
ms_size3 = []
ms_size4 = []
ms_wr_type = []
```

```python
for link in product_link4[:500]:
    driver.get(link)
    time.sleep(1.5)
    
    
    #Finding the Product price
    try:
        price_element=driver.find_element(By.XPATH,'//div[@class="prod-sp"]')
        if price_element:
            ms_p_price.append(price_element.text.replace('₹',''))
        else:
            ms_p_price.append('price not found')
    except Exception as e:
        ms_p_price.append('----')
        
    #Finding the Brand name
    try:
        brand_element=driver.find_element(By.XPATH,'//h2[@class="brand-name"]')
        if brand_element:
            ms_b_name.append(brand_element.text)
        else:
            ms_b_name.append('brand not found')
    except Exception as e:
        ms_b_name.append('----')
        
    #Finding the Product name
    try:
        pd_element=driver.find_element(By.XPATH,'//h1[@class="prod-name"]')
        if pd_element:
            ms_p_name.append(pd_element.text)
        else:
            ms_p_name.append('not found')
    except Exception as e:
        ms_p_name.append('----')

    #clicking the other information of the product
    try:
        info_element=driver.find_element(By.XPATH,'//ul[@class="prod-list"]/li/div[@class="other-info-toggle"]')
        info_element.click()
        time.sleep(2) 
    except Exception as e:
        print('not present')
        
    #Finding the Product code
    try:
        pdc_element=driver.find_element(By.XPATH,'//section[@class="prod-desc"]/h2[2]/ul/li/span')
        if pdc_element:
            ms_p_code.append(pdc_element.text.replace('Product Code:',''))
        else:
            ms_p_code.append('Code not found')
            
    except Exception as e:
        ms_p_code.append('----')
        
    #Finding the Commodity
    try:
        pt_element=driver.find_elements(By.XPATH,'//ul[@class="prod-list"]/li/div[@class="mandatory-list"]/div[@class="title"]')
        ms_p_type.append(pt_element[-1].text)
    except Exception as e:
        ms_p_type.append("----") 
       
    #fab type
    try:
        pd_element=driver.find_element(By.XPATH,'//ul[@class="prod-list"]/li[4]')
        if pd_element:
            ms_m_type.append(pd_element.text.replace('Product Code:',''))
        else:
            ms_m_type.append('----')
    except Exception as e:
        ms_p_type.append("----")
                
    #adding category
    ms_category.append('MEN')
    
    ms_occ.append('casual')
    
    ms_r_time.append('Easy 10 days return and exchange')
    
    ms_c_ori.append('India')
    
    # scrapping size 
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-swatch"][1]/div/span')
        ms_size1.append(size.text.strip())
    except Exception as e:
        ms_size1.append('-----')
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-swatch"][2]/div/span')
        ms_size2.append(size.text.strip())
    except Exception as e:
        ms_size2.append('-----')
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-swatch"][3]/div/span')
        ms_size3.append(size.text.strip())
    except Exception as e:
        ms_size3.append('-----')
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-swatch"][4]/div/span')
        ms_size4.append(size.text.strip())
    except Exception as e:
        ms_size4.append('-----')
    
    ms_wr_type.append('top wear')
    
    
    ms_col1.append('black')
    ms_col2.append('brown')
    ms_col3.append('White')
    ms_col4.append ('Grey')

    ms_reviews.append(random.randint(10,500))
```

```python
print("Length of ms_p_price:", len(ms_p_price))
print("Length of ms_b_name:", len(ms_b_name))
print("Length of ms_p_code:", len(ms_p_code))
print("Length of ms_category:", len(ms_category))
print("Length of ms_p_type:", len(ms_p_type))
print("Length of ms_p_name:", len(ms_p_name))

print("Length of ms_col1:", len(ms_col1))
print("Length of ms_col2:", len(ms_col2))
print("Length of ms_col3:", len(ms_col3))
print("Length of ms_col4:", len(ms_col4))

print("Length of ms_m_type:", len(ms_m_type))
print("Length of ms_reviews:", len(ms_reviews))

print("Length of ms_occ:", len(ms_occ))
print("Length of ms_r_time:", len(ms_r_time))
print("Length of ms_c_ori:", len(ms_c_ori))
print("Length of ms_size1:", len(ms_size1))
print("Length of ms_size2:", len(ms_size2))
print("Length of ms_size3:", len(ms_size3))
print("Length of ms_size4:", len(ms_size4))
print("Length of ms_wr_type:", len(ms_wr_type))
```

```python
data1 = pd.DataFrame({'Brand name' :ms_b_name, 'category' : ms_category, 'product type' : ms_p_type ,'product name' :ms_p_name, 'product code' :ms_p_code, 'price' :ms_p_price, 
                     'wear type' : ms_wr_type,"material type":ms_m_type,'colour 1' :ms_col1 , 'colour 2' :ms_col2 , 'colour 3' : ms_col3 , 'colour 4' : ms_col4, "reviews" : ms_reviews,
                      "occasion" : ms_occ,"return_ time" : ms_r_time,"country _origin" : ms_c_ori, 'size1' :ms_size1, 'size2' :ms_size2, 'size3' :ms_size3, 'size4' :ms_size4})
data1.to_csv('mdata1.csv', index = False)
```
