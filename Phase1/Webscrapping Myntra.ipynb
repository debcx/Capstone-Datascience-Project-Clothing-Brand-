```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
import time
warnings.filterwarnings("ignore")
```

# mens topwear

```python
driver = webdriver.Chrome()
driver.get('https://www.myntra.com/men-topwear')

search1 = driver.find_element(By.XPATH, '//ul[@class="categories-list"]/li[2]')
search1.click()
time.sleep(10)

#Scrapping the colours
for i in range(1,4):
    search1 = driver.find_element(By.XPATH, f'/html/body/div[2]/div/main/div[3]/div[1]/section/div/div[5]/ul/li[{i}]/label')
    search1.click()
    time.sleep(10)

# scrapping product links
product_links = []
for k in range(0,5):
    links = driver.find_elements(By.XPATH, '//ul[@class="results-base"]/li/a')
    for i in links:
        product_links.append(i.get_attribute('href'))
        
    driver.execute_script("window.scrollTo(0, 6000);")
    
    search1 = driver.find_element(By.XPATH, '//li[@class="pagination-next"]')
    search1.click()
    time.sleep(10)
    
p_price = []
b_name = []
p_code = []
category = []
p_como = []
p_name =[]

col1 = []
col2 = []
col3 = []
col4 = []
wr_type = []
m_type = []
reviews = []

occ = []
r_time = []
c_ori = []
size1 = []
size2 = []
size3 = []
size4 = []

for link in product_links:
    driver.get(link)
    time.sleep(2)  # Add a short delay to allow the page to load (adjust as needed)
    # Find the brand name 
    try:
        name_element = driver.find_element(By.XPATH, '//div[@class="pdp-price-info"]/h1[1]')
        if name_element:
            b_name.append(name_element.text)
        else:
            b_name.append('Name not found')
    except Exception as e:
        b_name.append('-----')
        
    # Find the product price
    try:
        price_element = driver.find_element(By.XPATH, '//span[@class="pdp-price"]')
        if price_element:
            p_price.append(price_element.text)
        else:
            p_price.append('Price not found')
    except Exception as e:
        p_price.append('-----')
        
    #product name 
    try:
        pr_name = driver.find_element(By.XPATH, '//div[@class="pdp-price-info"]/h1[2]')
        p_name.append(pr_name.text)
    except Exception as e:
        p_name.append('-----')
    
    try:
    # Loop through elements from index 2 to index 7
        for i in range(2, 10):
            head = driver.find_element(By.XPATH, f'//div[@class="index-tableContainer"]/div[{i}]/div[1]')
            if head.text == "Occasion":
                como = driver.find_element(By.XPATH, f'//div[@class="index-tableContainer"]/div[{i}]/div[2]').text
                p_como.append(como)
                break  # Exit the loop once "Occasion" is found
        else:
            p_como.append('-----')  # Append '-----' if "Occasion" is not found in any iteration
    except Exception as e:
        p_como.append('-----')
        
    # Find the product code element within the expanded details section
    try:
        code_element = driver.find_element(By.XPATH, '//div[@class="undefined supplier-desktopCodeSupplier"]/div[1]')
        if code_element:
            p_code.append(code_element.text)
        else:
            p_code.append('Code not found')
    except Exception as e:
        p_code.append('-----')
        
    #adding material type
    try:
        mt_name = driver.find_element(By.XPATH, '//div[@class="pdp-productDescriptorsContainer"]/div[3]/p')
        m_type.append(mt_name.text.strip())
    except Exception as e:
        m_type.append('-----')
        
    # adding reviews
    try:
        element = driver.find_element(By.XPATH, '//div[@class="index-countDesc"]')
        text = element.text
        number = ''.join(filter(str.isdigit, text))
        reviews.append(number)
    except Exception as e:
        reviews.append('------')
        
    #adding category
    category.append('male')
    
    #adding wear type 
    wr_type.append('top wear')
    
    try: #scrapping ocassion 
    # Loop through elements from index 2 to index 7
        for i in range(2, 10):
            head = driver.find_element(By.XPATH, f'//div[@class="index-tableContainer"]/div[{i}]/div[1]')
            if head.text == "Occasion":
                como = driver.find_element(By.XPATH, f'//div[@class="index-tableContainer"]/div[{i}]/div[2]').text
                occ.append(como)
                break  # Exit the loop once "Occasion" is found
        else:
            occ.append('-----')  # Append '-----' if "Occasion" is not found in any iteration
    except Exception as e:
        occ.append('-----')
        
    #scrapping return time
    try:
        r_type = driver.find_element(By.XPATH, '//div[@class="meta-container"]/div[3]/div')
        r_time.append(r_type.text.strip())
    except Exception as e:
        r_time.append('-----')
    
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-buttons-size-buttons"]/div[1]/div/button/p')
        size1.append(size.text.strip())
    except Exception as e:
        size1.append('-----')
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-buttons-size-buttons"]/div[2]/div/button/p')
        size2.append(size.text.strip())
    except Exception as e:
        size2.append('-----')
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-buttons-size-buttons"]/div[3]/div/button/p')
        size3.append(size.text.strip())
    except Exception as e:
        size3.append('-----')
    try:
        size = driver.find_element(By.XPATH, '//div[@class="size-buttons-size-buttons"]/div[4]/div/button/p')
        size4.append(size.text.strip())
    except Exception as e:
            size4.append('-----')
            
    #adding colours
    col1.append('Blue')
    col2.append('Navy Blue')
    col3.append('Grey')
    col4.append('White')
    
    #adding country of origin
    c_ori.append('india')
        
```

```python
list_lengths = {
    'p_price': len(p_price),
    'b_name': len(b_name),
    'p_code': len(p_code),
    'category': len(category),
    'p_como': len(p_como),
    'p_name': len(p_name),
    'wr_type': len(wr_type),
    'm_type': len(m_type),
    'reviews': len(reviews),
    'col1': len(col1),
    'col2': len(col2),
    'col3': len(col3),
    'col4': len(col4)
}

print("Length of each list:")
for key, value in list_lengths.items():
    print(f"{key}: {value}")
```

```python
data1 = pd.DataFrame({'Brand name' : b_name, 'category' : category, 'product type' : p_como,'product name' : p_name, 'product code' : p_code , 'price' :p_price, 
                      'wear type' : wr_type, 'material type' : m_type, "reviews" : reviews,'colour 1' : col1, 'colour 2' :col2, 'colour 3' : col3, 'colour 4' : col4,
                      "occasion" : occ, "return_ time" : r_time , "country _origin" : c_ori, 'size1' :size1, 'size2' :size2, 'size3' :size3, 'size4' :size4})
data1
```
